###NOTE : For Gender Coloumn Male = 1 & Female = 0

```python
df_new=df.drop(columns=['GENDER','AGE', 'SMOKING', 'SHORTNESS OF BREATH'])
df_new
```

```python
cn=df_new.corr()
cn
```

-> The correlation matrix shows that ANXIETY and YELLOW_FINGERS are correlated more than 50%. So, lets create a new feature combining them.

```python
df_new['ANXYELFIN']=df_new['ANXIETY']*df_new['YELLOW_FINGERS']
df_new
```

```python
#Splitting independent and dependent variables
X = df_new.drop('LUNG_CANCER', axis = 1)
y = df_new['LUNG_CANCER']
```

```python
from imblearn.over_sampling import ADASYN
adasyn = ADASYN(random_state=42)
X, y = adasyn.fit_resample(X, y)
```

```python
len(X)
```

```python
#Splitting data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(random_state=0)
lr_model.fit(X_train, y_train)
```

```python
y_lr_pred= lr_model.predict(X_test)
y_lr_pred
```

```python
#Model accuracy
from sklearn.metrics import classification_report, accuracy_score, f1_score
lr_cr=classification_report(y_test, y_lr_pred)
print(lr_cr)
```

This model is almost 97% accurate.

```python
# Make predictions on the existing dataset
predicted_probabilities = lr_model.predict_proba(X_test)

# The second column of the predicted_probabilities array contains the probability of the positive class
positive_class_probabilities = predicted_probabilities[:, 1]

# Print the probabilities
print(positive_class_probabilities)
```
